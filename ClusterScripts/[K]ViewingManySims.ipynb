```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

%load_ext autoreload
%autoreload 2

import numpy as np
from matplotlib import pyplot as plt

import sys
sys.path.append('..')
from BBH_kick_sim import Simulator

import pickle
import glob
```

## Uniform spin, No mass sort

```python
fileloc = './sp_U_sm_0/'
files = glob.glob(fileloc + '/*.pkl')
[print(f) for f in files];
```

```python
spUsm0_mass = []
spUsm0_spin = []
spUsm0_vel = []
spUsm0_LargestBHs = []
for fname in files:
    with open(fname,'rb') as f:
        params,seeds,all_gcs = pickle.load(f)
        
    #These files are BIG. Only do one at a time
    for gc in all_gcs:
        bhs = gc.ejected+gc.BHs
        spUsm0_mass.extend([b.m for b in bhs]) 
        spUsm0_spin.extend([b.s_mag for b in bhs])
        spUsm0_vel.extend([np.sqrt(np.sum(np.square(b.v))) for b in bhs])
        maxi = np.argmax([b.m for b in bhs])
        spUsm0_LargestBHs.append(bhs[maxi])
    del all_gcs
```

```python
n,_,_=plt.hist(spUsm0_mass,bins='auto',density=True)
plt.title('BH Final Mass distribution')
plt.xlabel('BH mass [M_sun]')
plt.show()
```

```python
n,_,_=plt.hist(spUsm0_spin,bins='auto',density=True)
plt.title('BH Final Spin distribution')
plt.xlabel('Dimentionless BH spin')
plt.show()
```

```python
n,_,_=plt.hist(spUsm0_vel,bins='auto',density=True)
plt.title('BH Final velocity')
plt.xlabel('Velocity [km/s]')
plt.show()
```

```python
maxi = np.argmax([b.m for b in spUsm0_LargestBHs])
spUsm0_LargestBHs[maxi].printTree(params=['m','s_mag','v_imag'])
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

%load_ext autoreload
%autoreload 2

import numpy as np
from matplotlib import pyplot as plt

import sys
sys.path.append('..')
from BBH_kick_sim import Simulator
```

# Uniform spin distribution

```python
params={
    'cluster_mass': 1e6, #Default=1e6
    'radius': 2, #Default=2
    'imf_alpha':2.25, #Default=2.35
    'min_bh_star':10, #Default=10
    'bh_mass_frac':0.5, #Default=0.5
    'use_mass_lookup':True, #Default=False
    'min_star':0.8, #Default=0.8
    'max_star':100, #Default=100
    'vel_thresh':0.1, #Default=0.1
    'rand_spin_type':'uniform', #Default='uniform'
}

unif_bhs = []
for i in range(10):
    sim = Simulator.Simulator(params,print_missing=True)
    sim.begin_sim(sort_mass_first=False,progress=10000)
    unif_bhs+=sim.GC.BHs+sim.GC.ejected
```

```python
unif_m = np.array([bh.m for bh in unif_bhs])
unif_s = np.array([bh.s_mag for bh in unif_bhs])
unif_v = np.array([np.sqrt(np.dot(bh.v,bh.v)) for bh in unif_bhs])

unif_bhs[np.argmax(unif_m)].printTree(params=['m','s','s_mag','v_imag'])
```

```python
plt.hist(unif_m,bins='auto',density=True)
plt.show()
```

```python
plt.hist(unif_s,bins='auto',density=True)
plt.title('Final Black hole spin distribution\nUniform initial spin')
plt.xlabel('Unitless final BH Spin')
plt.show()
```

```python
plt.hist(unif_v,bins='auto',density=True)
plt.show()
```

# Zero spin

```python
params={
    'cluster_mass': 1e6, #Default=1e6
    'radius': 2, #Default=2
    'imf_alpha':2.25, #Default=2.35
    'min_bh_star':10, #Default=10
    'bh_mass_frac':0.5, #Default=0.5
    'use_mass_lookup':True, #Default=False
    'min_star':0.8, #Default=0.8
    'max_star':100, #Default=100
    'vel_thresh':0.1, #Default=0.1
    'rand_spin_type':'zero', #Default='uniform'
}

zero_bhs = []
for i in range(10):
    sim = Simulator.Simulator(params,print_missing=True)
    sim.begin_sim(sort_mass_first=False,progress=10000)
    zero_bhs+=sim.GC.BHs+sim.GC.ejected
```

```python
zero_m = np.array([bh.m for bh in zero_bhs])
zero_s = np.array([bh.s_mag for bh in zero_bhs])
zero_v = np.array([np.sqrt(np.dot(bh.v,bh.v)) for bh in zero_bhs])

zero_bhs[np.argmax(zero_m)].printTree(params=['m','s','s_mag','v_imag'])
```

```python
plt.hist(zero_m,bins='auto',density=True)
plt.show()
```

```python
plt.hist(zero_s,bins='auto',density=True)
plt.title('Final Black hole spin distribution\nZero initial spin')
plt.xlabel('Unitless final BH Spin')
plt.vlines(0.68,0,7,colors='r')
plt.show()
```

```python
sorti = np.argsort(zero_s)
spinRange1 = [i for i in sorti if zero_s[i]>0.65]
spinRange2 = [i for i in sorti if zero_s[i]>0.7]

zero_bhs[spinRange1[4]].printTree(params=['m','s','s_mag','v_imag'])
zero_bhs[spinRange2[2]].printTree(params=['m','s','s_mag','v_imag'])
```

```python
plt.hist(zero_v,bins='auto',density=True)
plt.show()
```
